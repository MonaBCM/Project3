```python
# Dependencies
import pandas as pd
import re
from pprint import pprint
import json
```

```python
# Load in File from resources
# 'movies.csv'
data_file="movies.csv"
```

```python
# Read and display with pandas
data_file_df=pd.read_csv(data_file)
data_file_df.head()
```

```python
data_file_df.info()
```

```python
#Drop null values
data_file_df.dropna(inplace=True)
```

```python
data_file_df.info()
```

```python
data_file_df.loc[data_file_df['Month'] == '2008', 'Month'] = 'June'
data_file_df.loc[data_file_df['Month'] == '2014', 'Month'] = 'July'

data_file_df['Runtime'] = data_file_df['Runtime'].astype(float)
```

```python
data_file_df.info()
```

```python
conversion = {
    '$': 1.0,
    '€': 1.11,
    '£': 1.25,
    '₩': 0.00077,
    'C': 0.75,
    '₹': 0.012,
    '¥': 0.0075,
    'A': 0.9,
    'S': 0.097,
    'D': 0.15,
    'N': 0.097 
}
```

```python
#Convert all Budgets in different currencies to USD
for id, value in enumerate(data_file_df['Budget']):
    if str(data_file_df.iloc[id].Budget)[0] in conversion.keys():
        data_file_df.loc[data_file_df['Budget'] == value, 'Budget'] = int(re.sub("[^0-9]", "", value)) * conversion[data_file_df.iloc[id].Budget[0]]
data_file_df.head()
```

```python
#Drop records that has Unknown Budget and Income
data_file_df = data_file_df[(data_file_df['Budget']!='Unknown')&(data_file_df['Income']!='Unknown')]
data_file_df.head()
```

```python
data_file_df.info()
```

```python
data_file_df['Year'].value_counts()
```

```python
#remove $ from budget
for id, value in enumerate(data_file_df['Income']):
    if str(data_file_df.iloc[id].Income)[0] == '$':
        data_file_df.loc[data_file_df['Income'] == value, 'Income'] = int(re.sub("[^0-9]", "", value))
data_file_df.head()
```

```python
data_file_df.info()
```

```python
data_file_df['Budget'] = data_file_df['Budget'].astype(str).astype(float)
data_file_df['Income'] = data_file_df['Income'].astype(str).astype(float)
```

```python
data_file_df.head()
```

```python
data_file_df.to_csv("movies_cleaned.csv", index=False, header=True)
```

```python
#Combining Genre Column to List
data_file_df['Genre'] = [x.split(', ') for x in data_file_df['Genre']]
data_file_df.head()
```

```python
#Combining Country_of_origin Column to List
data_file_df['Country_of_origin'] = [x.split(', ') for x in data_file_df['Country_of_origin']]
data_file_df.head()
```

```python
#Combining Stars Column to List
data_file_df['Stars'] = [x.split(', ') for x in data_file_df['Stars']]
data_file_df.head()
```

```python
#Combining Directors Column to List
data_file_df['Directors'] = [x.split(', ') for x in data_file_df['Directors']]
data_file_df.head()
```

```python
#convert DF to json
data_file_df.to_json(r'output.json',orient='records')
```
