## Where are movies mostly filmed? 


```python
df_new = df[['Title', 'Filming_location']]
df_new
```

```python
country_counts = df['Filming_location'].value_counts()
country_counts
```

```python
# Create a new dataframe from the country_counts dictionary
df_country_counts = pd.DataFrame(list(country_counts.items()), columns=['Filming_location', 'Count'])
```

```python
# Write the new dataframe to a new csv file
df_country_counts.to_json('Filming_location.json', orient='records')
```

### Which country made the most movie? 

```python
df
```

```python
df_coc = df[['Title', 'Country_of_origin']]
df_coc
```

```python
Country_of_origin = df_coc.assign(Country_of_origin=df.Country_of_origin.str.split(", ")).explode("Country_of_origin")
Country_of_origin
```

```python
country_count = Country_of_origin['Country_of_origin'].value_counts()
```

```python
country_count
```

```python
df_country_count = pd.DataFrame(list(country_count.items()), columns=['Country_of_origin', 'Title'])
```

```python
df_country_count.to_json('Country_of_production.json', orient='records')
```

## AVG Runtime per year

## AVG budget and income 	

```python
df
```

```python
df_new2 = df[['Year', 'Runtime']]
df_new2 
```

```python
df_mean = df.groupby('Year', as_index=False)[['Runtime', 'Budget', 'Income']].mean(numeric_only=True)
```

```python
df_mean
```

```python
df_mean['Runtime'] = df_mean['Runtime'].astype(int)
```

```python
df_mean = df_mean.rename(columns={'Runtime': 'Avg_Runtime', 'Budget': 'Avg_Budget', 'Income': 'Avg_Income'})

df_mean
```

```python
df_mean.to_json('avgdata.json', orient='records')
```
